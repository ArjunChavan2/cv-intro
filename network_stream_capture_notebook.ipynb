```python
import cv2
from matplotlib import pyplot as plt
import lane_detection
import lane_following
```

```python
IP_ADDRESS = "192.168.1.102"
```

```python
vcap = cv2.VideoCapture(f"rtsp://{IP_ADDRESS}:8554/rovcam")

img = None

MAX_RETRY = 100
count = 0
while count < MAX_RETRY:
    count += 1
    # Obtain the frame
    ret, frame = vcap.read()

    # Check frame was received successfully
    if ret:
        # got a frame, close the cap and return the frame
        print(" got a frame ")
        img = frame
        vcap.release()

vcap.release()
```

```python
#plt.imshow(img)
#type(img)
img1 = cv2.imread("frame.jpg")
#cv2.imwrite("frame.jpg",img1)
plt.imshow(img1)
```

```python
lines = lane_detection.detect_lines(img1)
lanes = lane_detection.detect_lanes(lines)
center, slope = lane_following.get_lane_center(lanes)
print(lane_following.recommend_direction(center, slope))
plt.imshow(img1)

```
